```python
import os
import requests
```

```python
# mapping genres to Deezer's genre IDs
GENRES = {
    "pop": 132,
    "rock": 152,
    "classical": 98,
    "jazz": 129,
}

TRACKS_PER_GENRE = 110
```

```python
OUT_DIR = "./data"

os.makedirs(OUT_DIR, exist_ok=True)
for genre in GENRES:
    os.makedirs(os.path.join(OUT_DIR, genre), exist_ok=True)
```

```python
s = requests.Session()

for genre, genre_id in GENRES.items():
    print("downloading songs of genre:", genre)

    url = f"https://api.deezer.com/genre/{genre_id}/artists"
    artists_resp = s.get(url).json()

    if "data" not in artists_resp:
        print("No data found for this genre")
        break

    count = 0
    for artist in artists_resp["data"]:
        artist_id = artist["id"]
        top_tracks_url = f"https://api.deezer.com/artist/{artist_id}/top?limit=5"
        tracks_resp = s.get(top_tracks_url).json()

        if "data" not in tracks_resp:
            continue

        for track in tracks_resp["data"]:
            preview_url = track.get("preview")
            if not preview_url:
                continue

            track_title = track["title"].replace("/", "_")
            filename = os.path.join(OUT_DIR, genre, f"{track_title}.mp3")

            # Download preview
            r = s.get(preview_url)
            with open(filename, "wb") as f:
                f.write(r.content)

            count += 1

            if count >= TRACKS_PER_GENRE:
                break
        if count >= TRACKS_PER_GENRE:
                break
```
